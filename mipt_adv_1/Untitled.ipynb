Здесь происходят три вещи. Во-первых, каждый вход умножается на вес (на схеме обозначен <span style="color:red">красным</span>):  
$$ x_1 -> x_1 * w_1 $$  
$$ x_2 -> x_2 * w_2 $$

Затем все взвешенные входы складываются вместе со смещением b (на схеме обозначен <span style="color:green">зеленым</span>):

$$(x_1 * w_1) + (x_2 * w_2) + b $$

Наконец, сумма передается через функцию активации (на схеме обозначена <span style="color:orange">желтым</span>):

$$ y = f((x_1 * w_1) + (x_2 * w_2) + b)$$

Функция активации используется для подключения несвязанных входных данных с выводом, у которого простая и предсказуемая форма. Как правило, в качестве используемой функцией активации берется функция сигмоида:
$$ f(x) = \frac{1}{1+e^{-x}}$$

```python
import numpy as np


def sigmoid(x):
    # Наша функция активации: f(x) = 1 / (1 + e^(-x))
    return 1 / (1 + np.exp(-x))


class Neuron:
    def __init__(self, weights, bias):
        self.weights = weights ## веса нейрона
        self.bias = bias ## смещение
 
    def feedforward(self, inputs):
        # Вводные данные о весе, добавление смещения 
        # и последующее использование функции активации
        # Перемножение векторов и добавление смещения 
        total = np.dot(self.weights, inputs) + self.bias
        return sigmoid(total)
 
 
weights = np.array([0, 1])  # w1 = 0, w2 = 1
bias = 4  # b = 4
n = Neuron(weights, bias)
 
x = np.array([2, 3])  # x1 = 2, x2 = 3
print(n.feedforward(x))  # 0.9990889488055994
```

нейрон состоит из весов(равно количеству переменных), смещения(так же называется Байес) и функцией активации) по сути нейрон похож на линейную регрессию, тем что инициализирует веса для переменных, перемножает их и суммирует, после чего пропускает через функцию активации
